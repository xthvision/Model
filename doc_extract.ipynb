#### Reading and Cleaning the data

```python
# Import the module
from docx import *

document = Document('specs.docx')
fullText = ""
for para in document.paragraphs:
    fullText = fullText + para.text
    fullText += '\n'
print(type(fullText))
full_arr = fullText.split('\n')
full_arr
```

```python
lines = []
for i in range(len(full_arr)):
    if(full_arr[i]!=''):
        lines.append(full_arr[i])
lines
```

```python
import re
output = []
test_name = "Assay"
index = 0
for str in lines:
    
    index = index+1
    pattern = r"\d+ *\) *([A-Za-z ]+)"
    
    result = re.findall(pattern, str)
        
    for i in range(len(result)):
        r = result[i].strip()
        if(r == test_name):
            print(r)
            print("YES")
            print(index-1)
```
