```python
class Sample:
  def __init__(self,sample_name, name, assay):
    self.sample_name = sample_name
    self.name = name
    self.assay = assay
    
  def get_sample_name(self):
    return self.sample_name

  def get_name(self):
    return self.name

  def get_assay(self):
    return self.assay
```

```python
data = tables[0].df
rows = data.shape[0]
rows = rows-2
rows
```

```python
data_list = []
row_index = 1
for i in range(len(tables)):
    
    data = tables[i].df
    
    while(row_index != rows):

        sample_obj = Sample(data[1][row_index], data[2][row_index], data[18][row_index])
        data_list.append(sample_obj)
        row_index = row_index + 1;
        
            
        print(sample_obj.get_name())
        print(sample_obj.get_sample_name())
        print(sample_obj.get_assay())
```
