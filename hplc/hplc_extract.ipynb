```python
class Sample:
  def __init__(self,sample_name, name, assay_result):
    self.sample_name = sample_name
    self.name = name
    self.assay = assay_result
    
  def get_sample_name(self):
    return self.sample_name

  def get_name(self):
    return self.name

  def get_assay_result(self):
    return self.assay
```
